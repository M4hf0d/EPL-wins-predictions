Data Cleaning

````python
del matches["comp"]
del matches["notes"]
matches["date"] = pd.to_datetime(matches["date"])
matches["target"] = (matches["result"] == "W").astype("int")
````

Creating Simples Predictors

````python
matches["venue_code"] = matches["venue"].astype("category").cat.codes  # Convertinc venue to numeic : 0 is away, 1 is home 

matches["opp_code"] = matches["opponent"].astype("category").cat.codes # Opponent Codes  

matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")   # Hour column

matches["day_code"] = matches["date"].dt.dayofweek  

matches

````

Initial Random forest classifier Model : 
<br>
it can pick up non linearities in the data, like hour (19 is not necessarly better than 15)
<br>
Training test split on time series data, we have to ensure that we're training with data from the past and testing on the present 

````python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)

train = matches[matches["date"] < '2022-01-01']

test = matches[matches["date"] > '2022-01-01']

predictors = ["venue_code", "opp_code", "hour", "day_code"]

rf.fit(train[predictors], train["target"])

preds = rf.predict(test[predictors])

from sklearn.metrics import accuracy_score

error = accuracy_score(test["target"], preds)

error


````

````python
combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))

pd.crosstab(index=combined["actual"], columns=combined["predicted"])

````

````python
from sklearn.metrics import precision_score

precision_score(test["target"], preds)


````

##### Creating more predictors for better accuracy :
 Splitting by team and Computing rolling averages, see how a team did on past matches and use that to predict the result, if a team lost all their matches, might be expected that they will lose the next match also



````python
def rolling_average(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats =group[cols].rolling(3, closed= 'left')  # 3 game rolling average, closed left means it DOES NOT includes the current game
    group[new_cols] = rolling_stats.mean()
    group = group.dropna(subset=new_cols)
    return group
````

##### Applying rolling average on one tema "Man united" 

````python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"] # Goals for, goals against, shots, shots on target, distance, free kicks, penalties, penalty attempts 
new_cols = [f"{c}_rolling" for c in cols]


grouped_matches = matches.groupby("team")
group = grouped_matches.get_group("Manchester United").sort_values("date")
rolling_average(group, cols, new_cols)
````

````python
"""Applying this function to all teams"""
matches_rolling = matches.groupby("team").apply(lambda x: rolling_average(x, cols, new_cols))
matches_rolling.index
````

As you can see, we have team as an index which we do not need 

````python
matches_rolling = matches_rolling.reset_index(drop=True)
matches_rolling
````

Retraining our model with new predictors

````python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error

combined, error = make_predictions(matches_rolling, predictors + new_cols)
print("Error:",error)
````

````python
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)
combined.head(10)
````
